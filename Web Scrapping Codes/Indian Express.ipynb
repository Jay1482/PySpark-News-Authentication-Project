```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://indianexpress.com/section/cities/page/2/'
article = requests.get(url)
```

```python
soup = bs(article.content, 'html.parser')
```

```python
soup.find_all('h2', class_='title')
```

```python
title = []
for b in soup.find_all('h2', class_='title'):
    title.append(b.text)
title
```

```python
links = soup.find_all('h2', class_='title')
```

```python
# links
```

```python
time=soup.find_all('div' ,class_="date")
```

```python
timestamp = []
for b in time:
    timestamp.append(b.text)
timestamp
```

**Final Funtion**

```python
data = []
for l in links:
    href = l.find('a')
    link = href.get('href')
    url1 = link
    article1 = requests.get(url1)
    soup1 = bs(article1.content, 'html.parser')
    news_content_div = soup1.find('div', {'id': 'pcl-full-content'})
    time_content=soup1.find('span', {'itemprop':'dateModified'})
  
    if news_content_div is not None:
        news_content = news_content_div.find_all('p')
        
        news_para = []
        for paragraph in news_content:
            news_para.append(paragraph.get_text())
        data.append([time_content.text,l.text,link,news_para])
    else:
        print(f"No news content found for URL: {url1}")
```

```python
data
```

```python
df = pd.DataFrame(data,columns = ['Time','Title','Link','Description'])
```
