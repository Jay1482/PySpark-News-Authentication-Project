## Importing Webscrapping Library 

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
abc_url="https://abcnews.go.com/"
```

```python
article = requests.get(abc_url)
```

```python
soup = bs(article.content,'lxml')
```

```python
links=soup.find_all('div' , class_="ContentList__Item")
```

```python
title = []
for b in links:
    title.append(b.text)
title
```

```python
data = []
for l in links:
    href = l.find('a')
    # print(href)
    link = href.get('href')
    # print(link)
    data.append(link)

data 
```

```python

data = []
for l in links:
    href = l.find('a')
    link = href.get('href')
    url1 = link
    article1 = requests.get(url1)
    soup1 = bs(article1.content, 'html.parser')
    news_content_div = soup1.find('article' ,class_='xvlf ZRif TKoO eaKK ')
    time_content=soup1.find('div', class_="VZTD ")
    
    if news_content_div is not None:
      news_content = news_content_div.find_all('p')
      content = []
      for b in news_content_div:
        content.append(b.text)
      data.append([time_content.text,l.text,link,content])
    else:
        print(f"No news content found for URL: {url1}")
```

```python
# data
```

```python
df = pd.DataFrame(data,columns = ['Time','Title','Link','Description'])
```
